### 3 - Extração dos links de Download de Vídeos do Vimeo

- Precisamos pegar os links em 1080p, 720p e 540p para importar os vídeos para uma nova plataforma

```python
# para pegar o dicionario do vimeo, use:
from dic import dicionario_vimeo
import pprint  # Faz um print identado

# você precisa chegar em:
# videos = [
#     {'uri': video_uri, 'nome': nome_video, 'duracao': duracao, 'link540p': link540p, 'link720p': link720p, 'link1080p': link1080p},
# ]
informations = dicionario_vimeo["data"]
link540p = ""
link720p = ""
link1080p = ""
Videos = []

for item in informations:
    video_uri = item["uri"]
    video_nome = item["name"]
    video_duracao = item["duration"]
    for video in item["download"]:
        if video["rendition"] == "540p":
            link540p = video["link"]
        if video["rendition"] == "720p":
            link720p = video["link"]
        if video["rendition"] == "1080p":
            link1080p = video["link"]
    Videos.append(
        {
            "uri": video_uri,
            "nome": video_nome,
            "duracao": video_duracao,
            "link540p": link540p,
            "link720p": link720p,
            "link1080p": link1080p,
        }
    )

pprint.pprint(Videos)
```
