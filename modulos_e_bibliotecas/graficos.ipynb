# Exibindo Gráficos

```python
import matplotlib.pyplot as plt
```

```python
vendas_meses = [1500,1727,1350,999,1050,1027,1022,1500,2000,1433,2100,2762]
meses = ['Jan', 'Fev', 'Mar', 'Abr', 'Mai', 'Jun', 'Jul', 'Ago', 'Set', 'Out', 'Nov', 'Dez']

plt.plot(meses,vendas_meses)
plt.ylabel('Vendas')
plt.xlabel('Meses')
plt.axis([0, 12, 0, max(vendas_meses) + 500])
plt.show
```

### Mais edições de Gráfico com Matplot lib

```python
import numpy as np
```

- Linha

```python
vendas = np.random.randint(1000, 3000, 50)
meses = np.arange(1, 51)
plt.plot(meses, vendas)
plt.ylabel('Vendas')
plt.xlabel('Meses')
plt.axis([0, 52, 0, max(vendas) + 200])
plt.show
```

- Editando o gráfico de linha

```python
vendas = np.random.randint(1000, 3000, 50)
meses = np.arange(1, 51)
plt.plot(meses, vendas, '--' , color='red')
plt.ylabel('Vendas')
plt.xlabel('Meses')
plt.axis([0, 52, 0, max(vendas) + 200])
plt.show
```

- Gráfico de dispersão

```python
plt.scatter(meses, vendas)
plt.show()
```

- Gráfico de barras

```python
plt.bar(meses, vendas)
plt.show()
```

- Trabalhando com Mútiplos Gráficos no mesmo "plot"

```python
plt.figure(1,figsize=(20,5))
plt.subplot(1,3,1)
plt.plot(meses, vendas, '--' , color='red')
plt.subplot(1,3,2)
plt.scatter(meses, vendas)
plt.subplot(1,3,3)
plt.bar(meses, vendas)
plt.show()
```
